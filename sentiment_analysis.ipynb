```python
import graphlab as gl
```

```python
dataset=gl.SFrame.read_csv('datasets\\6.data_spell_corrected.csv',column_type_hints=[int,float,str,str,str],header=True)
```

```python
dataset.head()
```

```python
dataset['Reviews'][2]
```

```python
remove_word=['aur']
```

```python
dataset['Reviews']=gl.text_analytics.tokenize(dataset['Reviews'])
```

```python
dataset['Reviews']=gl.text_analytics.count_words(dataset['Reviews'])
```

```python
dataset['Reviews']=dataset['Reviews'].dict_trim_by_keys(gl.text_analytics.stopwords(), True)
dataset['Reviews']=dataset['Reviews'].dict_trim_by_keys(remove_word, True)
#dataset['Reviews']=dataset['Reviews'].dict_trim_by_keys(remove_wordset, True)
```

### Classification using logistic regression

```python
# ignore all 3* reviews
#data=data[data['Rating']!=3]
# positive sentiment is 4* or 5* reviews
dataset['sentiment']=dataset['Rating']>3
```

```python
train_data,test_data=dataset.random_split(.7,seed=0)
```

```python
sentiment_model=gl.logistic_classifier.create(train_data,target='sentiment',features=['Reviews'],validation_set=test_data)
```

```python
sentiment_model.evaluate(test_data,metric='roc_curve')
```
