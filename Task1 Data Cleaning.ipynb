# Data Cleaning for NLP Zomato Dataset

```python
# basic libraries required in this project
import pandas as pd
```

```python
# read the data file
mydata = pd.read_excel("Datasets/dataset1.xlsx")
print "The size of the actual dataset :", len(mydata)
```

```python
# capture the row index for which number of words in the reveiws are less than 10
x = mydata["Review"]
# list of indices to remove
newx=[]
for i in range(0,len(x)-1):
    if(x[i]):
        if len(x[i].split(" "))<10:
            newx.append(i)
    else:
        newx.append(i)
```

```python
print "Sample review to be removed :"
print x[newx[0]]
print "Length of the review :" ,len(x[448].split(" "))

# delete the reviews from the main dataset having less than 10 words
for itm in newx:
    mydata = mydata.drop(mydata.index[[itm]])

print "Size of dataset after removing reviews having less than 10 words :", len(mydata)
```

```python
# remove the redundant entries
mydata = pd.DataFrame.drop_duplicates(mydata)
print "Size after removing redundant entries :", len(mydata)
```

```python
# Create a Pandas Excel writer using XlsxWriter as the engine.
writer = pd.ExcelWriter('Datasets/pandas_output.xlsx', engine='xlsxwriter')

# Convert the dataframe to an XlsxWriter Excel object.
mydata.to_excel(writer, sheet_name='Sheet1')

# Close the Pandas Excel writer and output the Excel file.
writer.save()
```
